```python
df["Price"]=df["Price"].str.replace("$","")
```

```python
df["Installs"]=df["Installs"].str.replace("+","")
```

```python
df["Installs"]=df["Installs"].str.replace(",","")
```

```python
df = df[df['Installs'] != '']
```

```python
df["Installs"]=df["Installs"].astype(int)
```

```python
df["Last Updated"] = pd.to_datetime(df["Last Updated"], format='%B %d, %Y', errors='coerce')
```

```python
df.info()
```

```python
ratings=df.pivot_table(values="Rating",index="Category",aggfunc="mean")
pd.options.display.float_format = '{:.1f}'.format
```

```python
ratings
```

```python
df1=df[df["Rating"].isna()]
```

```python
df1=pd.merge(df1,ratings,on="Category",how="left")
df1.drop(columns=["Rating_x"],inplace=True)
```

```python
df.rename(columns={'Rating_y': 'Rating'}, inplace=True)
```

```python
df2=df[df["Rating"].notnull()]
```

```python
df = pd.concat([df1, df2], ignore_index=True)
```

```python
df.drop(columns=["Rating_y"],inplace=True)
```

### Question 1: Which is the most used Android version?

```python
android_version=df.pivot_table(values="App",index="Android Ver",aggfunc="count")
```

```python
aa=android_version.sort_values(by="App",ascending=False).head(10)
aa
```

```python
aa.plot(kind="bar",color="green")
plt.title("Top 10 Android Versions.")
plt.xlabel("Android Version")
plt.ylabel("Count of Apps")
plt.legend()
```

##### The most used Android version based on the apps is 4.1 and up with 2451 apps supporting it.

### Question 3: Which is the most installed category of mob app from this dataset?

```python
installed=df.pivot_table(values="Installs",index="Category",aggfunc="sum",margins=True,margins_name="Total")
```

```python
installed["Percentage"]=(installed["Installs"]/installed.loc["Total","Installs"])*100
installed=installed.sort_values(by="Percentage",ascending=False).head(11)
installed
```

```python
installed.tail(10).plot(kind="bar",color="purple")
plt.title("TOP 10 App Categories.")
plt.xlabel("Category")
plt.ylabel("No. of Apps")
plt.legend()
```

##### The most installed category is GAME {20.93%}, followed by COMMUNICATION {19.48}. This shows most of people either are interested in games or communication based apps, implying huge competition between the apps of these categories.
